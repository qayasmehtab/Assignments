# Q1 Write a program that accepts a string from user. Your program should count and display number of vowels in that string.

```python
def count_vowels(input_string):
    """
    Counts the number of vowels in a given string.

    Args:
        input_string: The string to be checked for vowels.

    Returns:
        The total count of vowels (a, e, i, o, u) in the string.
    """
     # Define the vowels
    vowels = "aeiou"
    
    # Initialize a counter for the vowels
    vowel_count = 0

    # Iterate through each character in the string
    for char in input_string:
        # Check if the character is a vowel
        if char in vowels:
            vowel_count += 1
            
    return vowel_count

# Get input from the user
user_input = input("Please enter a string: ")

# Call the function and store the result
total_vowels = count_vowels(user_input)

# Display the result to the user
print(f"The number of vowels in the string is: {total_vowels}")
```

```python
def count_vowels(input_string):
    """
    Counts lowercase and uppercase vowels separately in a given string.

    Args:
        input_string: The string to be analyzed.

    Returns:
        A dictionary containing the counts for 'lowercase_vowels' and 'uppercase_vowels'.
    """
    # Initialize separate counters for lowercase and uppercase vowels
    lowercase_vowels = 0
    uppercase_vowels = 0
    
    # Define the sets of vowels
    LOWERCASE_VOWELS = "aeiou"
    UPPERCASE_VOWELS = "AEIOU"

    # Iterate through each character in the string
    for char in input_string:
        if char in LOWERCASE_VOWELS:
            lowercase_vowels += 1
        elif char in UPPERCASE_VOWELS:
            uppercase_vowels += 1
            
    # Return the counts in a dictionary for clarity
    return {
        'lowercase_vowels': lowercase_vowels,
        'uppercase_vowels': uppercase_vowels
    }

# Get input from the user
user_input = input("Please enter a string: ")

# Call the function and get the results
vowel_counts = count_vowels(user_input)

# Display the results to the user
print("\n--- Vowel Count Results ---")
print(f"Original String: '{user_input}'")
print(f"Lowercase Vowels (a,e,i,o,u): {vowel_counts['lowercase_vowels']}")
print(f"Uppercase Vowels (A,E,I,O,U): {vowel_counts['uppercase_vowels']}")
print("---------------------------")
```

# Q2 Write a program that reads a string from keyboard and display:
* The number of uppercase letters in the string
* The number of lowercase letters in the string
* The number of digits in the string
* The number of whitespace characters in the string

```python
# A program to count uppercase, lowercase, digits, and whitespace characters in a string.

def count_characters(input_string):
    """
    Counts different types of characters in a string.

    Args:
        input_string (str): The string to be analyzed.

    Returns:
        tuple: A tuple containing the counts in the order:
               (uppercase, lowercase, digits, whitespace).
    """
    
    # Initialize all counters to zero
    uppercase_count = 0
    lowercase_count = 0
    digit_count = 0
    whitespace_count = 0
    
    # Loop through each character in the string
    for char in input_string:
        # Check if the character is an uppercase letter
        if char.isupper():
            uppercase_count += 1
        # Check if it's a lowercase letter
        elif char.islower():
            lowercase_count += 1
        # Check if it's a digit (0-9)
        elif char.isdigit():
            digit_count += 1
        # Check if it's a whitespace character (like a space, tab, or newline)
        elif char.isspace():
            whitespace_count += 1
            
    # Return all the counts
    return (uppercase_count, lowercase_count, digit_count, whitespace_count)

# --- Main part of the program ---

# Prompt the user for input
user_input = input("Please enter a string: ")

# Call the function and unpack the returned tuple into separate variables
upper, lower, digits, spaces = count_characters(user_input)

# Display the results
print("\n--- Character Counts ---")
print(f"Number of uppercase letters: {upper}")
print(f"Number of lowercase letters: {lower}")
print(f"Number of digits: {digits}")
print(f"Number of whitespace characters: {spaces}")
```

# Q3 Write a Python program that accepts a string from user. Your program should create and display a new string where the first and last characters have been exchanged.

```python
# This program swaps the first and last characters of a string.

def swap_characters(input_string):
    """
    Swaps the first and last characters of a given string.

    Args:
        input_string (str): The string to modify.

    Returns:
        str: The new string with the first and last characters exchanged.
             Returns the original string if it has less than 2 characters.
    """

    # Check if the string has at least two characters.
    # If not, it's impossible to swap the first and last characters, so we return the original string.
    if len(input_string) < 2:
        return input_string

    # Get the first character of the string.
    # In Python, string indexing starts at 0.
    first_char = input_string[0]
    
    # Get the last character of the string.
    # The index -1 is a shortcut to get the last character of any string.
    last_char = input_string[-1]
    
    # Get all the characters in the middle of the string.
    # This slicing starts from index 1 (the second character) and goes up to, but not including, the last character.
    middle_chars = input_string[1:-1]
    
    # Combine the characters to form the new string.
    # The last character comes first, followed by the middle characters, and then the first character.
    new_string = last_char + middle_chars + first_char
    
    # Return the newly created string.
    return new_string

# --- Main part of the program ---

# Prompt the user to enter a string.
user_string = input("Please enter a string: ")

# Call the function with the user's string and save the result.
swapped_string = swap_characters(user_string)

# Display the final output.
print(f"The original string was: {user_string}")
print(f"The new string is: {swapped_string}")
```

# Q4 Write a Python program that accepts a string from user. Your program should create a new string in reverse of first string and display it.

```python
# This program takes a string from the user and reverses it.

def reverse_string(input_string):
    """
    Reverses the order of characters in a string.

    This function uses Python's slicing feature to create a new string
    that is a reversed copy of the original string.
    
    Args:
        input_string (str): The string to be reversed.

    Returns:
        str: A new string with the characters in reverse order.
    """
    
    # The key to this program is string slicing with a step of -1.
    # [start:stop:step]
    # Leaving 'start' and 'stop' empty means 'from the beginning to the end'.
    # A 'step' of -1 tells Python to go backwards.
    return input_string[::-1]

# --- Main part of the program ---

# Get a string from the user via the keyboard.
user_string = input("Please enter a string to reverse: ")

# Call the function to reverse the string.
reversed_string = reverse_string(user_string)

# Display the original and the new, reversed string.
print(f"The original string is: {user_string}")
print(f"The reversed string is: {reversed_string}")
```

# Q5 Write a Python program that accepts a string from user. Your program should create a new string by shifting one position to left.

```python
# This program shifts a string one position to the left.

def left_shift(input_string):
    """
    Shifts a string one position to the left.

    This function takes the first character of the string and moves it
    to the end. For example, "python" becomes "ythonp".
    
    Args:
        input_string (str): The string to be shifted.

    Returns:
        str: The new string shifted one position to the left.
             Returns the original string if it has less than 2 characters.
    """
    
    # First, check if the string has at least two characters.
    # If not, shifting is not meaningful, so we return the original string.
    if len(input_string) < 2:
        return input_string
        
    # Get all characters from the second character to the end of the string.
    # The slice [1:] starts at index 1 and goes all the way to the end.
    shifted_part = input_string[1:]
    
    # Get the very first character of the string.
    first_char = input_string[0]
    
    # Combine the two parts to form the new string.
    # We add the first character to the end of the shifted part.
    new_string = shifted_part + first_char
    
    # Return the newly created string.
    return new_string

# --- Main part of the program ---

# Prompt the user to enter a string.
user_string = input("Please enter a string: ")

# Call the function to shift the string to the left.
shifted_string = left_shift(user_string)

# Display the original and the new, shifted string.
print(f"The original string is: {user_string}")
print(f"The new, shifted string is: {shifted_string}")
```

# Q6 Write a program that asks the user to input his name and print its initials. Assuming that the user always types first name, middle name and last name and does not include any unnecessary spaces. 

```python
# This program extracts the initials from a user's full name.

def get_initials(full_name):
    """
    Takes a full name and returns the initials.

    This function assumes the name contains a first, middle, and last name,
    separated by single spaces. It splits the string into words and takes
    the first character of each to form the initials.

    Args:
        full_name (str): The user's full name.

    Returns:
        str: The initials in uppercase.
    """
    
    # Split the full name string into a list of words.
    # The split() method breaks the string at each space.
    # For "John David Smith", this creates the list: ['John', 'David', 'Smith']
    name_parts = full_name.split()
    
    # Get the first letter of each word and convert it to uppercase.
    first_initial = name_parts[0][0].upper()
    middle_initial = name_parts[1][0].upper()
    last_initial = name_parts[2][0].upper()
    
    # Combine the initials with a period after each.
    initials = f"{first_initial}.{middle_initial}.{last_initial}."
    
    # Return the formatted initials.
    return initials

# --- Main part of the program ---

# Ask the user to input their full name.
user_name = input("Please enter your full name (First, Middle, Last): ")

# Get the initials by calling the function.
user_initials = get_initials(user_name)

# Display the result to the user.
print(f"Your initials are: {user_initials}")
```

# Q7 A palindrome is a string that reads the same backward as forward. For example, the words dad, madam and radar are all palindromes. Write a programs that determines whether the string is a palindrome.
* Note: do not use reverse() method


```python
# This program determines if a string is a palindrome.

def is_palindrome(input_string):
    """
    Checks if a given string is a palindrome without using the reverse() method.

    A palindrome is a string that reads the same backward as forward.
    The function handles different cases by normalizing the string (e.g.,
    removing spaces and making it lowercase) before comparison.

    Args:
        input_string (str): The string to be checked.

    Returns:
        bool: True if the string is a palindrome, False otherwise.
    """
    
    # Step 1: Normalize the string
    # We'll create a new, clean string by removing spaces and converting all
    # characters to lowercase. This makes the check case-insensitive.
    clean_string = "".join(input_string.split()).lower()
    
    # Step 2: Manually reverse the string
    # We create an empty string to build the reversed version.
    reversed_string = ""
    
    # Loop backwards through the clean string.
    # We use a 'for' loop with 'range' to go from the last character's index
    # (len(clean_string) - 1) down to the first index (0).
    for i in range(len(clean_string) - 1, -1, -1):
        # Add each character to the reversed_string
        reversed_string += clean_string[i]
    
    # Step 3: Compare the strings
    # The final step is to check if the cleaned string is the same as the reversed string.
    return clean_string == reversed_string

# --- Main part of the program ---

# Get a string from the user.
user_string = input("Enter a string to check if it's a palindrome: ")

# Call the function to check for palindrome status.
if is_palindrome(user_string):
    print(f"'{user_string}' is a palindrome!")
else:
    print(f"'{user_string}' is NOT a palindrome.")
```

# Q8 Write a program that display following output:
* SHIFT
* HIFTS
* IFTSH
* FTSHI
* TSHIF
* SHIFT

```python
# This program displays a string that is left-shifted in each line.

def display_shifted_string(s):
    """
    Displays a string that is progressively shifted one position to the left
    in each line, returning to the original string at the end.

    Args:
        s (str): The string to be shifted and displayed.
    """
    
    # Store the original string so we can stop the loop when we return to it.
    original_string = s
    
    # Print the first line, which is the original string.
    print(s)
    
    # Start a loop to generate the shifting pattern.
    # The loop will continue as long as the current string 's' is not
    # the same as the original string.
    while True:
        # Shift the string one position to the left.
        # This is done by taking all characters from the second one onwards (s[1:])
        # and adding the first character (s[0]) to the end.
        s = s[1:] + s[0]
        
        # Print the newly shifted string.
        print(s)
        
        # Check if the shifted string has returned to the original.
        # If it has, we break out of the loop.
        if s == original_string:
            break

# --- Main part of the program ---

# The string to be shifted.
my_string = "SHIFT"

# Call the function to display the output.
display_shifted_string(my_string)
```

# Q9 Write a program in python that accepts a string to setup a passwords. Your entered password must meet the following requirements: 

```python
# This program validates a password based on a set of common security rules.

def validate_password(password):
    """
    Checks if the given password meets standard security requirements.

    Args:
        password (str): The password string to be validated.

    Returns:
        tuple: A tuple containing a boolean result (True if valid, False if not)
               and a list of reasons for failure.
    """
    
    # Define the requirements
    min_length = 8
    
    # Initialize a list to store failure reasons
    failure_reasons = []
    
    # Flags to check for character types
    has_uppercase = False
    has_lowercase = False
    has_digit = False
    has_special_char = False
    
    # List of allowed special characters
    special_chars = "!@#$%^&*()"
    
    # Rule 1: Check password length
    if len(password) < min_length:
        failure_reasons.append("Password must be at least 8 characters long.")
        
    # Rule 2, 3, 4, 5: Check for character types
    for char in password:
        if 'A' <= char <= 'Z':
            has_uppercase = True
        elif 'a' <= char <= 'z':
            has_lowercase = True
        elif '0' <= char <= '9':
            has_digit = True
        elif char in special_chars:
            has_special_char = True
            
    if not has_uppercase:
        failure_reasons.append("Password must contain at least one uppercase letter.")
    if not has_lowercase:
        failure_reasons.append("Password must contain at least one lowercase letter.")
    if not has_digit:
        failure_reasons.append("Password must contain at least one digit.")
    if not has_special_char:
        failure_reasons.append(f"Password must contain at least one special character from: {special_chars}")
        
    # Check if any reasons for failure were added
    if failure_reasons:
        return (False, failure_reasons)
    else:
        return (True, [])

# --- Main part of the program ---

# Get a password from the user
user_password = input("Enter a password to validate: ")

# Call the function to validate the password
is_valid, reasons = validate_password(user_password)

# Display the validation result
print("\n--- Validation Result ---")
if is_valid:
    print("Success! Your password meets all the requirements.")
else:
    print("Failed. Your password did not meet the following requirements:")
    for reason in reasons:
        print(f"- {reason}")
```
