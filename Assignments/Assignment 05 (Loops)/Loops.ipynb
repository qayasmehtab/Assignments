### Q1. Write a Python program to print the numbers from 1 to 10 using a for loop.

```python
for num in range(1,11):
    print(num)
```

### Q2. Write a Python program to print the numbers from 20 to 1 using a while loop.

```python
num = 20
while num >= 1:
    print(num)
    num -= 1
```

### Q4. Write a program that prompts the user to enter a number n and prints all the numbers from 1 to n.

```python
n = int(input("Enter a number:"))
print(f"number from 1, {n}:")
for i in range(1, n + 1):
    print(i)
```

### Q5. Write a program that prompts the user to enter a number n, and then prints all the odd numbers between 1 and n.

```python
n = int(input("Enter a number: "))
print(f"Odd numbers between 1 and {n}:")
for i in range(1, n + 1):
    if i % 2 != 0:
        print(i)
```

### Q7. Write a program that takes a number n as input from the user and generates the firstn terms of the series formed by squaring the natural numbers.

* 1 4 9 16 25 36

### Q11. Write a program that asks the user for a positive integer value. The program should calculate the sum of all the integers from 1 up to the number entered. For example, if the user enters 20, the loop will find the sum of 1, 2, 3, 4, ... 20.

```python
n = int(input("Enter a positive integer: "))

total = 0
for i in range(1, n + 1):
    total = total + i
    print(f"Adding {i}: total = {total}")

print(f"\nThe sum of all integers from 1 to {n} is: {total}")
```

### Q12. write a program that takes a positive integer N as input and calculates the sum of the reciprocals of all numbers from 1 up to N. The program should display the final sum.
* Output of the program should be like:
* Enter a positive integer: 5
* The sum of reciprocals from 1 to 5 is: 2.28

```python
n = int(input("Enter a positive integer: "))

total = 0
for i in range(1, n + 1):
    total = total + 1/i

print(f"The sum of reciprocals from 1 to {n} is: {total:.2f}")
```

### Q13. Write a program that prompts the user to enter a number and repeats this process 5 times. The program should accumulate the numbers entered and then display the final
* running total.
* Sample Output:
* Enter a number: 10
* Enter a number: 15
* Enter a number: 35
* Enter a number: 40
* Enter a number: 50
* The final running total is: 150

```python
total = 0
for i in range(5):
    num = int(input("Enter a number: "))
    total = total + num

print(f"The final running total is: {total}")
```

### Q14. Write a program that prompts the user to enter a positive integer and calculates its factorial. The factorial of a positive integer 'n' is denoted as 'n!' and is calculated by multiplying all the integers from 1 to 'n' together. For example, the factorial of 5
* (denoted as 5!) is calculated as 1 x 2 x 3 x 4 x 5.
* The program should display the factorial value if the input is a positive number, or display a message stating that the factorial does not exist for negative numbers.
* Additionally, for an input of zero, the program should output that the factorial of 0 is 1.

```python
n = int(input("Enter a positive integer: "))

if n < 0:
    print("Factorial does not exist for negative numbers.")
elif n == 0:
    print("The factorial of 0 is 1")
else:
    factorial = 1
    for i in range(1, n + 1):
        factorial = factorial * i
    print(f"The factorial of {n} is: {factorial}")
```

### Q15. Write a Python program that prompts the user to enter a base number and anexponent, and then calculates the power of the base to the exponent. The programshould not use the exponentiation operator (**) or the math.pow() function. Theprogram should handle both positive and negative exponents.

```python
base = int(input("Enter the base number: "))
exponent = int(input("Enter the exponent: "))

result = 1

for i in range(abs(exponent)):
    result = result * base

if exponent < 0:
    result = 1 / result

print(f"{base} raised to the power {exponent} is: {result}")
```
