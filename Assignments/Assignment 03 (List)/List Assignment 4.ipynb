* Q3 Find and display the largest number of a list without using built-in function max(). Your program
should ask the user to input values in list from keyboard

```python
def reverse_with_loop(s):
    reversed_s = ""
    for char in s: 
        reversed_s = char + reversed_s
    return reversed_s

word = input("Enter a word or phrase: ")
result = reverse_with_loop(word)

print(f"Original: {word}")
print(f"Reversed: {result}")
```

*Q4 Write a program that rotates the element of a list so that the element at the first index moves to the
second index, the element in the second index moves to the third index, etc., and the element in the last
index moves to the first index.

```python
def rotate_list(data_list):
    return data_list[-1:] + data_list[:-1]
user_input_input = input("Enter elements separated by commas (Cat,Dog,Bird): ")
original_list = [item.strip() for item in user_input_input.split(',') if item.strip()]

if not original_list:
    print("The list is empty. Nothing to rotate.")
else:
    rotated_list = rotate_list(original_list)
    print("\n--- List Rotation Results ---")
    print(f"Original List: {original_list}")
    print(f"Rotated List:  {rotated_list}")
```

* Q5 Write a program that input a string and ask user to delete a given word from a string.

```python
def user_delete_word():
    sentence = input("Enter a sentence: ")
    word_to_remove = input("Enter the word to delete (e.g., 'the'): ")
    new_sentence = sentence.replace(" " + word_to_remove + " ", " ")
    new_sentence = new_sentence.replace(word_to_remove + " ", "")
    new_sentence = new_sentence.replace(" " + word_to_remove, "") 
    new_sentence = new_sentence.strip() 
    print("\n--- Output ---")
    print(f"Original: {sentence}")
    print(f"Removed '{word_to_remove}'")
    print(f"Result:   {new_sentence}")

user_delete_word()
```

* Q6 Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It
should print the date in the form March 12, 2021

```python
def convert_date_format():
    month_names = {
        '01': 'January', '02': 'February', '03': 'March', '04': 'April',
        '05': 'May', '06': 'June', '07': 'July', '08': 'August',
        '09': 'September', '10': 'October', '11': 'November', '12': 'December'
    }
    date_str = input("Enter a date in MM/DD/YYYY format (e.g., 03/12/2024): ").strip()
    if len(date_str) != 10 or date_str[2] != '/' or date_str[5] != '/':
        print("\nError: Date must be exactly 10 characters long and use the MM/DD/YYYY format.")
        return
    try:
        month_num = date_str[0:2] # MM
        day_num = date_str[3:5]   # DD
        year_num = date_str[6:10] # YYYY
    except IndexError:
        print("\nError: Could not parse the date parts. Check your formatting.")
        return
    month_name = month_names.get(month_num)
    if month_name is None:
        print(f"\nError: Invalid month number '{month_num}'. Must be between 01 and 12.")
        return
    try:
        clean_day = str(int(day_num))
    except ValueError:
        print(f"\nError: Invalid day number '{day_num}'.")
        return
    new_date_format = f"{month_name} {clean_day}, {year_num}"
    print("\n--- Date Conversion Results ---")
    print(f"Original Input: {date_str}")
    print(f"Converted Date: **{new_date_format}")
convert_date_format()
```

* Q8 Find the sum of each row of matrix of size m x n. For example for the following matrix output will be
like this.
- Sum of row 1 = 32
- Sum of row 1 = 31
- Sum of row 1 = 63

```python
def calculate_row_sums(matrix):
        print("\n--- Row Summation Results ---")
        for i, row in enumerate(matrix):
            row_total = sum(row)
            print(f"Sum of row {i + 1} = {row_total}")

my_matrix = [
    [10, 5, 12, 5],  # Row 1 (Sum = 32)
    [8, 11, 4, 8],   # Row 2 (Sum = 31)
    [20, 15, 18, 10] # Row 3 (Sum = 63)
]
print("Matrix used for calculation:")
for row in my_matrix:
    print(row)
calculate_row_sums(my_matrix)
```

* Q9 Write a program to add two matrices of size n x m.

```python
def add_matrices(matrix_a, matrix_b):
    rows_a = len(matrix_a)
    cols_a = len(matrix_a[0]) if rows_a > 0 else 0
    
    rows_b = len(matrix_b)
    cols_b = len(matrix_b[0]) if rows_b > 0 else 0

    if rows_a != rows_b or cols_a != cols_b:
        print("\nError: Cannot add matrices. Dimensions must match (n x m).")
        return None

    result_matrix = [[0 for j in range(cols_a)] for i in range(rows_a)]
    for i in range(rows_a):
        for j in range(cols_a):
            result_matrix[i][j] = matrix_a[i][j] + matrix_b[i][j]

    return result_matrix
A = [
    [10, 5, 2],
    [3, 8, 1]
]

B = [
    [4, 1, 9],
    [7, 2, 6]
]

print("Matrix A:")
for row in A:
    print(row)

print("\nMatrix B:")
for row in B:
    print(row)

Sum_AB = add_matrices(A, B)

print("--- Matrix Addition Results (A + B) ---")
if Sum_AB:
    for row in Sum_AB:
        print(row)
    print("\nAddition successful because matrices A and B have the same size (2x3).")
```

* Q10 Write a program to multiply two matrices

```python
def multiply_matrices(matrix_a, matrix_b):
    rows_a = len(matrix_a)
    cols_a = len(matrix_a[0]) if rows_a > 0 else 0
    
    rows_b = len(matrix_b)
    cols_b = len(matrix_b[0]) if rows_b > 0 else 0
    
    if cols_a != rows_b:
        print("\nError: Cannot multiply matrices. Columns of first matrix must equal rows of second matrix.")
        print(f"Required: {cols_a} == {rows_b}")
        return None

    result_matrix = [[0 for j in range(cols_b)] for i in range(rows_a)]
    
    for i in range(rows_a):
        for j in range(cols_b):
            for k in range(cols_a):   # <-- missing loop added
                result_matrix[i][j] += matrix_a[i][k] * matrix_b[k][j]

    return result_matrix


A = [
    [1, 2, 3],
    [4, 5, 6]
]

B = [
    [7, 8],
    [9, 10],
    [11, 12]
]

print("Matrix A (2x3):")
for row in A:
    print(row)

print("\nMatrix B (3x2):")
for row in B:
    print(row)

Product_AB = multiply_matrices(A, B)

print("\n--- Matrix Multiplication Results (A x B) ---")
if Product_AB:
    for row in Product_AB:
        print(row)
    print("\nMultiplication successful. Resulting matrix size is 2x2.")
```
